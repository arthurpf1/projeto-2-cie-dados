```python
muito_relevante={}     #DICIONARIO MUITO RELEVANTE
for i in range(len(df2.resultado)):
    for palavra in df2.Treinamento[i]:
        if df2.resultado[i] == 'muito relevante':
            if palavra not in muito_relevante:
                muito_relevante[palavra]=1
            else:
                muito_relevante[palavra]+=1

relevante = {}     #DICIONARIO RELEVANTE
for i in range(len(df2.resultado)):
    for palavra in df2.Treinamento[i]:
        if df2.resultado[i] == 'relevante':
            if palavra not in relevante:
                relevante[palavra]=1
            else:
                relevante[palavra]+=1

neutro = {}     #DICIONARIO NEUTRO
for i in range(len(df2.resultado)):
    for palavra in df2.Treinamento[i]:
        if df2.resultado[i] == 'neutro':
            if palavra not in neutro:
                neutro[palavra]=1
            else:
                neutro[palavra]+=1
irrelevante = {}     #DICIONARIO IRRELEVANTE
for i in range(len(df2.resultado)):
    for palavra in df2.Treinamento[i]:
        if df2.resultado[i] == 'irrelevante':
            if palavra not in irrelevante:
                irrelevante[palavra]=1
            else:
                irrelevante[palavra]+=1

muito_irrelevante = {}     #DICIONARIO MUITO IRRELEVANTE
for i in range(len(df2.resultado)):
    for palavra in df2.Treinamento[i]:
        if df2.resultado[i] == 'muito irrelevante':
            if palavra not in muito_irrelevante:
                muito_irrelevante[palavra]=1
            else:
                muito_irrelevante[palavra]+=1
```

```python
from collections import Counter                
lista_palavras=[muito_relevante,relevante,neutro,irrelevante,muito_irrelevante]
c=Counter()
for i in lista_palavras:
    c.update(i)
```

```python
pmrelevante={}     #DICIONARIO PROBABILIDADE PALAVRA MUITO RELEVANTE
prelevante={}     #DICIONARIO PROBABILIDADE PALAVRA RELEVANTE
pneutro={}     #DICIONARIO PROBABILIDADE PALAVRA NEUTRO
pirrelevante={}     #DICIONARIO PROBABILIDADE PALAVRA IRRELEVANTE
pmirrelevante={}     #DICIONARIO PROBABILIDADE PALAVRA MUITO IRRELEVANTE
for i in c.keys():
    if i in muito_relevante.keys():
        x=(muito_relevante[i] + 1)/(sum(c.values()) + sum(muito_relevante.values()))
        pmrelevante[i]=x
    if i in relevante.keys():
        x=(relevante[i] + 1)/(sum(c.values()) + sum(relevante.values()))
        prelevante[i]=x
    if i in neutro.keys():
        x=(neutro[i] + 1)/(sum(c.values()) + sum(neutro.values()))
        pneutro[i]=x
    if i in irrelevante.keys():
        x=(irrelevante[i] + 1)/(sum(c.values()) + sum(irrelevante.values()))
        pirrelevante[i]=x
    if i in muito_irrelevante.keys():
        x=(muito_irrelevante[i] + 1)/(sum(c.values()) + sum(muito_irrelevante.values()))
        pmirrelevante[i]=x
```
